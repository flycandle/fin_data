```python
plt.plot(df_merged.index, df_merged['UNRATE'])
plt.plot(df_merged.index, df_merged['M2SL'])
```

```python
fig,ax = plt.subplots()
ax.plot(df_merged.index, df_merged['UNRATE'],color="blue",marker="o")
ax2=ax.twinx()
ax2.plot(df_merged.index, df_merged['M2SL'],color="red")
fig.savefig('.jpg',format='jpeg')
```
