```python
df1 = data1.to_frame(name="FPCPITOTLZGUSA")
df2 = data2.to_frame(name='UNRATE')
```

```python
import pandas as pd
df_merged = pd.concat([df1, df2], axis=1)
```

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
plt.plot(df_merged.index, df_merged['FPCPITOTLZGUSA'])
plt.plot(df_merged.index, df_merged['UNRATE'])
```

```python
fig,ax = plt.subplots()
ax.plot(df_merged.index, df_merged['FPCPITOTLZGUSA'],color="blue",marker="o")
ax2=ax.twinx()
ax2.plot(df_merged.index, df_merged['UNRATE'],color="red")
fig.savefig('.jpg',format='jpeg')
```
